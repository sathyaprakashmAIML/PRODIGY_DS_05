```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("BigML_Dataset_5fde1fade84f94757d00135a.csv")
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Race']=df['Race'].fillna("")
```

```python
sns.histplot(df['Fatalities in crash'], bins=20,kde=True)
plt.title('Distribution of Fatalities in Crash')
plt.show()
```

```python
state_crashes = df['State'].value_counts()
state_crashes.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Crashes by State')
plt.ylabel('Number of Crashes')
plt.xlabel('State')
plt.show()
```

```python
gender_dist = df['Gender'].value_counts()
gender_dist.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title('Gender Distribution in Crashes')
plt.show()
```

```python
sns.countplot(x='Alcohol Results', hue='Injury Severity', data=df)
plt.title('Crash Severity Based on Alcohol Results')
plt.show()
```

```python
monthly_crashes = df.groupby('Crash Date.month').size()
monthly_crashes.plot(kind='line', marker='o')
plt.title('Monthly Trend of Crashes')
plt.ylabel('Number of Crashes')
plt.xlabel('Month')
plt.show()
```

```python
sns.countplot(x='Drug Involvement', data=df)
plt.title('Drug Involvement in Crashes')
plt.show()
```
